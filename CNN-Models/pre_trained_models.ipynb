# Pre-trained Model

A model that has been trained on a large dataset for a specific task and is available for reuse in other related tasks.

Pretrained models are commonly used in deep learning, especially for tasks such as image classification, object detection, natural language processing, and more.

# ResNet-50

ResNet-50 is a convolutional neural network (CNN) architecture that belongs to the ResNet (Residual Network) family. ResNet-50 is characterized by its depth, featuring 50 layers, and is known for its ability to train very deep neural networks effectively.

## What problems ResNets solve?


ResNets address the vanishing gradient problem by facilitating the flow of gradients through skip connections backwards from later layers to initial filters.

## What is Vanishing Gradient problem?

Vanishing Gradient happens when the network is too deep, the gradients from where the loss function is calculated easily shrink to zero after several applications of the chain rule.

Because of which the parameters (weights/bais) never updating its values and therefore, no training (learning) is being performed.

image.png

# Key features of ResNet-50:

* **Residual Blocks**: ResNet-50 introduces the concept of residual blocks, which allow for the training of very deep networks without suffering from the vanishing gradient problem.

  - In a residual block, the input to a convolutional layer is added to its output, allowing the network to learn residual mappings rather than directly learning the desired mappings. This helps with the flow of gradients during backpropagation and facilitates the training of very deep networks.

* **Bottleneck Architecture**: ResNet-50 uses a bottleneck architecture in its residual blocks to reduce computational complexity.

  - Each residual block consists of three convolutional layers:
    - a 1x1 convolutional layer (to reduce the dimensionality of the input),
    - a 3x3 convolutional layer (to capture spatial information),
    - and another 1x1 convolutional layer (to restore the dimensionality of the input).

  - This bottleneck architecture helps reduce the number of parameters and computational cost while preserving representational power.
  
* **Skip Connections**: ResNet-50 includes skip connections (also known as shortcut connections or identity mappings) that skip one or more layers in the network.

   - These connections enable information from earlier layers to bypass certain intermediate layers and be directly propagated to later layers.

  - Skip connections help mitigate the degradation problem that can occur in very deep networks and facilitate the training of deeper networks.

* **Architecture**: ResNet-50 consists of a stack of residual blocks, followed by global average pooling and a fully connected layer for classification.

  - It has a total of 50 layers, including convolutional layers, batch normalization layers, activation functions (ReLU), max-pooling layers, and fully connected layers.

* **Prediction Layer**: The final layer produces the output predictions with softmax activation function to predict 1000 Classes.

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights = "imagenet")
```

## Predicting Cat

```python
img_path = "/content/cat.jpg"
img = image.load_img(img_path, target_size = (224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x = preprocess_input(x)
```

```python
prediction = model.predict(x)
print("Predicted Image is: ", decode_predictions(prediction, top = 5)[0])
```
