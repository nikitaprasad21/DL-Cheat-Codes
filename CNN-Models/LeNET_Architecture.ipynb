# LeNet-5

A classic CNN architecture proposed by Yann LeCun et al. in 1998 for handwritten digit recognition tasks.

It is one of the pioneering CNN architectures and played a significant role in popularizing deep learning for computer vision tasks.

LeNet-5 consists of convolutional layers, pooling layers, and fully connected layers.

## The LeNet-5 architecture:

* **Input**: 32x32 grayscale images
* **Convolutional Layer 1**: 6 filters (5x5), followed by max-pooling
* **Convolutional Layer 2**: 16 filters (5x5), followed by max-pooling
* **Fully Connected Layer 1**: 120 units
* **Fully Connected Layer 2**: 84 units
* **Output Layer**: 10 units (for digit classification)




```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
from keras import Sequential
```

image.png

The LeNet-5 architecture was designed for grayscale images of handwritten digits, such as those found in the MNIST dataset.

 It achieved state-of-the-art performance on the MNIST dataset at the time of its publication and demonstrated the effectiveness of CNNs for image recognition tasks.

```python
model = Sequential()

# Layer 1
model.add(Conv2D(6, kernel_size = (5,5), padding = "valid", activation = "tanh", input_shape = (32,32,1)))
model.add(MaxPooling2D(pool_size = (2,2), strides = 2, padding= "valid"))

# Layer 2
model.add(Conv2D(16, kernel_size = (5,5), padding = "valid", activation = "tanh",))
model.add(MaxPooling2D(pool_size = (2,2), strides = 2, padding= "valid"))

model.add(Flatten())

# Layer 3
model.add(Dense(120, activation = "tanh"))

# Layer 4
model.add(Dense(80, activation = "tanh"))

# Layer 5
model.add(Dense(10, activation = "softmax"))
```

```python
model.summary()
```

LeNet-5 has a relatively small number of parameters compared to modern CNN architectures, making it computationally efficient and suitable for training on CPUs.

However, its design principles and architectural components have influenced the development of more advanced CNN architectures used in deep learning today.
