# Bidirectional Recurrent Neural Networks (RNNs):

 RNN architecture processing sequences in both forward and backward directions, capturing bidirectional context for more comprehensive sequence understanding.

image.png

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense, Bidirectional
```

```python
# Load the IMDb dataset
(train_input,train_target),(test_input,test_target) = imdb.load_data(num_words=10000)

# Pad sequences to have the same length
train_input = pad_sequences(train_input, maxlen=100, padding = "post", truncating = "post")
test_input = pad_sequences(test_input, maxlen=100, padding = "post", truncating = "post")
```

## Bi-directional RNNs

Standard RNN architecture with bidirectional processing, commonly used for tasks like sequence labeling and sentiment analysis.

```python
# Define the RNN model
model_1 = Sequential([
    Embedding(input_dim = 10000, output_dim = 32, input_length=100), # Embedding layer to convert words to vectors
    # RNN layer with 5 units
    Bidirectional(SimpleRNN(5)),
    Dense(1, activation='sigmoid')          # Output layer for binary classification
])

model_1.summary()
```

```python
# Compile the model
model_1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

image.png

```python
# Train the model
history_1 = model_1.fit(train_input, train_target, epochs=5, batch_size=32, validation_data=(test_input,test_target))
```

## Bi-directional LSTMs (Bi-LSTM):

RNN architecture with bidirectional LSTM units, capable of capturing long-term dependencies in both forward and backward directions.

image.png

```python
# Define the LSTM model
model_2 = Sequential([
    Embedding(input_dim = 10000, output_dim = 32, input_length=100), # Embedding layer to convert words to vectors
    # LSTM layer with 5 units
    Bidirectional(LSTM(5)),
    Dense(1, activation='sigmoid')          # Output layer for binary classification
])

model_2.summary()
```

```python
# Compile the model
model_2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history_2 = model_2.fit(train_input, train_target, epochs=5, batch_size=32, validation_data=(test_input,test_target))
```

## Bi-directional GRUs

 (Bi-GRU): RNN architecture with bidirectional GRU units, offering a simpler alternative to Bi-LSTM with comparable performance in many tasks.


image.png

```python
# Define the GRU model
model_3 = Sequential([
    Embedding(input_dim = 10000, output_dim = 32, input_length=100), # Embedding layer to convert words to vectors
    # GRU layer with 5 units
    Bidirectional(GRU(5)),
    Dense(1, activation='sigmoid')          # Output layer for binary classification
])

model_3.summary()
```

```python
# Compile the model
model_3.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history_3 = model_3.fit(train_input, train_target, epochs=5, batch_size=32, validation_data=(test_input,test_target))
```



### Advantages:

* **Enhanced Context**: Capture bidirectional context, allowing for better understanding of sequential data and improved performance on tasks like sentiment analysis and sequence labeling (NER or POS).
* **Improved Performance**: More comprehensive context integration often leads to higher accuracy in tasks that benefit from bidirectional information processing.

### Disadvantages:

* **Increased Complexity**: Higher computational cost and model complexity compared to unidirectional RNNs due to processing sequences in both directions.
* **Longer Training Time**: Bidirectional processing requires more parameters and longer training times, especially in deep architectures, potentially leading to longer training convergence.
